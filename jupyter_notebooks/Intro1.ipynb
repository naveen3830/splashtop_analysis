```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import os
import time
import re

def clean_text(text):
    # Remove extra whitespace and newlines
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

def extract_content(url):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')

        # Remove script and style elements
        for script in soup(["script", "style"]):
            script.decompose()

        # Get text
        text = soup.get_text()

        # Clean text
        cleaned_text = clean_text(text)

        return cleaned_text
    except Exception as e:
        print(f"Error extracting content from {url}: {str(e)}")
        return None

def save_content(url_id, content):
    filename = f"{url_id}.txt"
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(content)
    print(f"Saved content to {filename}")

# Read the CSV file
df = pd.read_csv('D:\Lead Walnut\links.csv')

# Create a directory to store the text files
os.makedirs('extracted_content', exist_ok=True)
os.chdir('extracted_content')

# Extract content for each URL and save to a text file
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']

    print(f"Processing {url_id}: {url}")

    content = extract_content(url)
    if content:
        save_content(url_id, content)

    # Add a small delay to avoid overwhelming the server
    time.sleep(1)

print("Extraction complete!")
```
